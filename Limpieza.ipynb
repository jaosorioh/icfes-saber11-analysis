```python
dtypes_sql = {'str':'VARCHAR(255)','int64':'INT','float64':'REAL','datetime64[us]':'DATE'}
def getTypes(cols, prefix, df):    
    types = {}
    for col in cols:
        clean_col = col.replace(prefix, "")
        types[clean_col] = str(df[col].dtype)
    return types
    
def getSQLTable(table, prefix, cols, df):
    types = getTypes(cols, prefix, df) 
    SQL_table = f'''CREATE TABLE IF NOT EXISTS {table}( 
id INT AUTO_INCREMENT PRIMARY KEY'''
        
    for col, dtype in types.items():
        SQL_table += f",\n{col.lower()} {dtypes_sql[dtype]}"
    SQL_table += '\n);'
    return SQL_table
    
```

```python
tables_dict = [{
    'name': 'estudiantes',
    'prefix': 'ESTU_',
    'cols': estu_selected_cols},
               {
    'name': 'familias',
    'prefix': 'FAMI_',
    'cols': fami_selected_cols}
              ]
for table in tables_dict:
   print(getSQLTable(table["name"], table["prefix"], table["cols"], df_ICFES))
```
