Agrupando las columnas en diferentes categorías

```python
estu_cols = [col for col in df_ICFES.columns if 'ESTU_' in col]
fami_cols = [col for col in df_ICFES.columns if 'FAMI_' in col]
cole_cols = [col for col in df_ICFES.columns if 'COLE_' in col]
result_cols = [col for col in df_ICFES.columns if not col.startswith(('COLE_', 'ESTU_', 'FAMI_'))]
```

De estas columnas, seleccionaremos solo las más importantes.

NSE significa nivel socio económico, 
TIENEETNIA - solo es un sí/no pero no se específica la ETNIA, por lo que no es significativo.
TIPODOCUMENTO, ESTUDIANTE, GENERACION-E, ESTADOINVESTIGACION no son relevantes. 

ESTU_NACIONALIDAD, ESTU_GENERO, ESTU_CONSECUTIVO ESTU_FECHANACIMIENTO, ESTU_PAIS_RESIDE, ESTU_DEPTO_RESIDE, ESTU_MCPIO_RESIDE, ESTU_NSE_INDIVIDUAL, ESTU_NSE_ESTABLECIMIENTO

serán las columnas de interés

```python
estu_selected_cols = ["ESTU_CONSECUTIVO", "ESTU_NACIONALIDAD", "ESTU_GENERO","ESTU_FECHANACIMIENTO","ESTU_PAIS_RESIDE", "ESTU_DEPTO_RESIDE", "ESTU_MCPIO_RESIDE", "ESTU_NSE_INDIVIDUAL", "ESTU_NSE_ESTABLECIMIENTO"]
```

```python
df_ICFES[estu_selected_cols].isna().sum()
```

Como los nulos de GENERO, DEPTO_RESIDE, NSE_ESTABLECIMIENTO, MCPIO_RESIDE son pocos, se eliminan, mientras la columna ESTU_NSE_INDIVIDUAL tiene una población significativa.

```python
df_ICFES.dropna(subset=['ESTU_NSE_ESTABLECIMIENTO','ESTU_GENERO','ESTU_DEPTO_RESIDE', 'ESTU_MCPIO_RESIDE'], inplace=True)
```

```python
df_ICFES[estu_selected_cols].isna().sum()
```

```python
df_ICFES[estu_selected_cols].info()
```

```python
df_ICFES["ESTU_FECHANACIMIENTO"] = pd.to_datetime(df_ICFES['ESTU_FECHANACIMIENTO'], format='%Y %b %d')
```

Conversión a SQL:

```python
dtypes_sql = {'str':'VARCHAR(255)','int64':'INT','float64':'REAL','datetime64[us]':'DATETIME'}

estud_table = f'CREATE TABLE  IF NOT EXISTS ESTUDIANTE\n ({estu_selected_cols[0]} {dtypes_sql[str(df_ICFES[estu_selected_cols[0]].dtype)]} PRIMARY KEY'

for col in estu_selected_cols[1:]:
    estud_table += f',\n {col} {dtypes_sql[str(df_ICFES[col].dtype)]}'

estud_table += ');'

print(estud_table)
```
